```python
def wrangle(filename):
    df = pd.read_csv(filename)
    df['Title']=df['Name'].str.split(',',expand=True).iloc[:,1].str.split('.').str[0]
    df['Last Name']=df['Name'].str.split(',',expand=True).iloc[:,0]
    last_name_count = df['Name'].str.split(',',expand=True).iloc[:,0].value_counts()
    last_name_count.reset_index().rename(columns={0:'Last Name'},inplace=True)
    df = pd.merge(df,last_name_count,left_on='Last Name',right_index=True)
    df['Cabin No'] = df['Cabin'].str.split().str.len()
    df['Cabin No'] = df['Cabin No'].fillna(0)
    df['Cabin No'] = df['Cabin No'].astype(int)
    df.drop(columns=['Name','Cabin','Ticket','PassengerId'],inplace=True)
    return df
```

```python
def wrangle(filename):
    df = pd.read_csv(filename)
    df['date'] = pd.to_datetime(df['date'])
    df['time'] = pd.to_datetime(df['time'])
    df['year'] = df['date'].dt.year
    df['month'] = df['date'].dt.month
    df['day'] = df['date'].dt.day
    df['quarter'] = df['date'].dt.quarter
    df['weekend'] = df['date'].dt.day_of_week.apply(lambda x : True if x > 4 else  False)
    df['hour'] =  df['time'].dt.hour
    df['minute'] =  df['time'].dt.minute
    df['night'] = df['time'].dt.hour.apply(lambda x : True if x > 19 or x < 5 else  False)
    df['crime_path'] = df['location_description'].str.split(' ').str[-1]
    df['street_name']=df['location_description'].str.split(' ').str[1]
    df.drop(['id','date','time','location_description'],axis=1,inplace=True)
    return df
# drop date column
```
